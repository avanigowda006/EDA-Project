# 6.Understanding Variables
The descriptions of the variables based on the netflix_titles dataset:
#1️. show_id:
Description: Unique identification number assigned to each movie or TV show in the dataset. It helps in distinguishing one record from another.
Data Type: Categorical (ID)
#2️. type
Description: Specifies whether the content is a Movie or a TV Show. It helps in categorizing and analyzing content separately.
Data Type: Categorical
#3️. title
Description: Name of the movie or TV show available on Netflix. It is used to identify and search for specific content.
Data Type: Text
#4️. director
Description: Name of the director(s) who directed the movie or TV show. Some entries may contain multiple names or missing values.
Data Type: Text
#5️. cast
Description: List of main actors and actresses featured in the content. It may contain multiple names separated by commas.
Data Type: Text
#6. country
Description: Country where the movie or TV show was produced. Some entries may contain multiple countries.
Data Type: Categorical
#7. date_added
Description: The date on which the content was added to Netflix platform. It helps in analyzing content addition trends over time.
Data Type: Date
#8. release_year
Description: The year in which the movie or TV show was originally released. It is useful for analyzing content distribution by year.
Data Type: Numerical
#9. rating
Description: Indicates the content rating such as TV-MA, PG-13, etc. It helps users understand age suitability.
Data Type: Categorical
#10. duration
Description: Duration of the movie in minutes or number of seasons for TV shows. The format varies depending on the content type.
Data Type: Categorical
#11. listed_in
Description: Genre or category of the movie or TV show (e.g., Drama, Comedy). It helps in content classification.
Data Type: Categorical
#12. description
Description: Short summary describing the storyline of the movie or TV show. It provides a brief overview of the content.
Data Type: Text

| Column Name  | Description            | Data Type   |
| ------------ | ---------------------- | ----------- |

| type         | Movie or TV Show       | Categorical |
| title        | Name of the content    | Text        |
| director     | Director name          | Text        |

| country      | Country of production  | Categorical |
| date_added   | Date added to Netflix  | Date        |
| release_year | Year of release        | Numerical   |
| rating       | Age rating             | Categorical |

#8.Data Wrangling

8.1 Column Handling

```python
# Rename columns
df.columns = df.columns.str.lower()

# Drop unnecessary columns
df.drop(['show_id', 'description'], axis=1, inplace=True)
```

8.2 Feature Creation

```python
#list of all column names in the dataset
df.columns
```

8.3 Date and Time Feature Extraction

```python
#Converts the date_added column from text (object) to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
#Creates a new column year_added containing only the year part of date_added
df['year_added'] = df['date_added'].dt.year
#Creates a new column month_added containing the month number (1–12).
df['month_added'] = df['date_added'].dt.month
#Creates a new column day_added containing the day of the month (1–31)
df['day_added'] = df['date_added'].dt.day
#Creates a new column quarter_added containing the quarter (1–4) of the year
df['quarter_added'] = df['date_added'].dt.quarter
```

```python
#Selects only the 3 columns: date_added, year_added, and month_added
#head() Displays the first 5 rows of the selected columns.
df[['date_added','year_added','month_added']].head()
```

8.4 Data Type Conversion

```python
## Convert 'type' column to category to save memory and speed up analysis
df['type'] = df['type'].astype('category')
#Convert 'rating' column to category to save memory and speed up analysis
df['rating'] = df['rating'].astype('category')
```

#9.Data Cleaning

9.1 Duplicate Handling

```python
# Count the number of duplicate rows in the dataset
duplicates = df.duplicated().sum()
print("Duplicate Rows:", duplicates)
# Remove all duplicate rows from the dataset
df = df.drop_duplicates()
```
